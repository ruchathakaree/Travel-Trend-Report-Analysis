```python
packages_df['Travel Date'] = pd.to_datetime(packages_df['Travel Date'])
# Convert the 'Travel Date' column in the travel package dataset to a datetime format
```

```python
# Convert the 'Destination' and city column to lowercase and remove spaces
packages_df['Destination'] = packages_df['Destination'].str.lower().str.strip()
cities_df['City'] = cities_df['City'].str.lower().str.strip()
```

```python
packages_df.head()
```

```python
cities_df.head()
```

```python
# Merge the travel package dataset (packages_df) with the city details dataset (cities_df)
# The merge is done on the 'Destination' column from packages_df and the 'City' column from cities_df
# 'left' join is used, meaning all records from packages_df will be retained, and matching records from cities_df will be added
merged_df = packages_df.merge(cities_df, left_on='Destination', right_on='City', how='left')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Get the top 10 most frequent destinations from the merged dataset
top_destinations = merged_df['Destination'].value_counts().head(10)
```

```python
# Create a bar plot to visualize the top 10 travel destinations
sns.barplot(x=top_destinations.index, y=top_destinations.values)
plt.xticks(rotation=45,fontsize=10)
plt.title("Top 10 Travel Destinations",fontsize=14)
plt.show
```

-The bars on the graph represent the most popular destinations. These are typically based on the number of travel packages, bookings, or popularity scores.

-By comparing the lengths of the bars, you can easily see which destinations are more in demand than others.

```python
# Extract the month from the 'Travel Date' column and create a new 'Month' column in the merged dataset
merged_df['Month'] = merged_df['Travel Date'].dt.month
```

```python
# Group the merged dataset by the 'Month' column and count the number of travel packages for each month
monthly_trends = merged_df.groupby("Month")['Package Name'].count()
```

```python
# Create a line plot to visualize the seasonal trends in monthly booking volume
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_trends.index, y=monthly_trends.values, marker='o', linewidth=2.5)
plt.xticks(range(1,13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title("Seasonal Travel Trends: Monthly Booking Volume", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Bookings", fontsize=12)
plt.grid(True)
plt.show()
```

-The months where the line peaks indicate the busiest travel periods (highest number of bookings).

-These peaks could correspond to holiday seasons, summer vacations, or major festivals when more people tend to travel.

-The seasonal trend line provides a clear view of when people prefer to travel.

-Peak months highlight high-demand travel periods, while dips indicate budget-friendly, less crowded times

```python
# Convert the 'Per Person Price' column to numeric, forcing any invalid entries to NaN (coerce)
packages_df['Per Person Price'] = pd.to_numeric(packages_df['Per Person Price'], errors='coerce')
df = packages_df.dropna(subset=['Per Person Price', 'Destination']) 
```

```python
# Group the dataset by 'Destination' and calculate the average 'Per Person Price' for each destination
# Then sort the results in descending order to find the highest average prices
avg_price_per_destination = df.groupby("Destination")['Per Person Price'].mean().sort_values(ascending=False)
```

```python
budget_trips = df[df['Per Person Price'] <4500]
```

```python
budget_trips['Month'] = budget_trips['Travel Date'].dt.month
avg_price_budget_monthly = budget_trips.groupby('Month')['Per Person Price'].mean()
```

```python
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
df.columns = df.columns.str.title()   # Capitalize column names (if needed)
print(df.columns)
```

```python
# Filter the data for budget packages only
budget_df = df[df['Package Type'] == 'Budget']

# Group by 'Month' and calculate the average price for budget packages
avg_budget_price_per_month = budget_df.groupby('Month')['Per Person Price'].mean()
plt.figure(figsize=(10,6))
# Plot the results as a line plot to visualize the best time for budget-friendly travel
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
avg_budget_price_per_month.plot(kind='line', marker='o', color='green')
plt.title('Best Time to Book Budget-Friendly Trips')
plt.xticks(range(1, 13), months) 
plt.xlabel('Month')
plt.ylabel('Average Price for Budget Packages')
plt.grid()
plt.show() 
```

-Best months to book budget trips: Likely in January, September, and early November.
    
-Most expensive months: Summer (June-August) and holidays (December, March-April).

-Travelers can save money by booking in low-price months or avoiding peak seasons when demand spikes.
    
-Travel businesses can use this data to adjust pricing and promotional strategies for maximum revenue.

```python
package_type_avg_price = packages_df.groupby('Package Type')['Per Person Price'].mean().sort_values(ascending=False)
```

```python
# Plot comparison using bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=package_type_avg_price.index, y=package_type_avg_price.values)
plt.title('Average Price for Luxury vs Budget Travel', fontsize=14)
plt.xlabel('Package Type', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.show()
```

-Luxury travel is significantly more expensive but provides premium services and exclusivity.

-Budget travel remains the preferred choice for cost-conscious travelers, offering affordability with fewer premium features.

-Understanding price differences can help travel companies target the right audience and optimize package offerings.

-Consumers should balance cost vs. experience based on their preferences and travel goals.

```python
flight_type_avg_price = df.groupby('Flight Stops')['Per Person Price'].mean().sort_values(ascending=False)
```

```python
# Create a bar plot with flight types (Direct vs Layover) on the x-axis and their average prices on the y-axis
plt.figure(figsize=(8,5))
sns.barplot(x=flight_type_avg_price.index, y=flight_type_avg_price.values)
plt.title("Average Price Comparison: Direct Flights vs Layovers", fontsize=14)
plt.xlabel("Flight Stops (Direct vs Layover)", fontsize=12)
plt.ylabel("Average Price (Per Person)", fontsize=12)
plt.grid(True)
plt.show()
```

-Typically, direct flights tend to be more expensive than flights with layovers because they are more convenient and save time.
